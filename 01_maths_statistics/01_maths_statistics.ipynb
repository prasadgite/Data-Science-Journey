load and view the dataset


```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url)
df.head()
```

Basic Info and Descriptive Statistics

```python
#shape and data type
print(df.shape)
print(df.dtypes)
#summary stats for numeric columns
df.describe()
```

Compure Descriptive Stats Column_Wise

```python
# Mean Age and Fare
print("Mean Age:", df['Age'].mean())
print("Mean Fare:", df['Fare'].mean())

# Median and Mode
print("Median Age:", df['Age'].median())
print("Mode Embarked:", df['Embarked'].mode()[0])

# Standard Deviation
print("Standard Deviation (Fare):", df['Fare'].std())

# IQR for Age
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
iqr = q3 - q1
print("IQR for Age:", iqr)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset again if needed
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Set seaborn style
sns.set(style="darkgrid")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"].dropna(),bins=30,kde=True,color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=df, palette='pastel')
plt.title('Passenger Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```
